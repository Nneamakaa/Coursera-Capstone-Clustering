```python
# Import Libraries 
import pandas as pd
import numpy as np
import requests # library to handle requests
```

Using Beautiful soup Library for web scrapping 

```python
#download the data
webdata = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(webdata,'lxml')
```

```python
# get the first table on page-as postal codes are in first table 
My_table = soup.find_all('table')[0] 
```

```python
# extract first row of table for headers
column_headers = [th.getText() for th in soup.findAll('tr')[0].findAll('th')]

# removing new lline character 
column_headers[-1] = column_headers[-1].strip()
column_headers
```

```python
# skip the first header row 
data_rows = soup.findAll('table')[0]('tr')[1:]  
```

```python
# extract the data in table 

Toronto_data = []  # create an empty list to hold all the data

for i in range(len(data_rows)): # for each table row
    data_row = []  # create an empty list for each pick/player

    # for each table data element from each table row
    for td in data_rows[i].findAll('td'):        
        # get the text content and append to the data_row 
        data_row.append(td.getText())        

    # then append each pick/player to the player_data matrix
    Toronto_data.append(data_row)
Toronto_data
```

```python
# convert into dataframe 

df = pd.DataFrame(Toronto_data , columns = column_headers  )
```

```python
# remove newline character
df.replace('\n',"",regex=True,inplace = True)
```

```python
df
```

```python
#Cleaning od Data
#remove 'Not assigned' from Borough column
df = df[df.Borough != 'Not assigned']

#replace not assigned -Neighbourhood cells with corresponding Borough name
df['Neighbourhood'].replace('Not assigned',df.Borough,inplace= True)
```

```python
#group data for common postcodes
df1 = df.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ','.join(x))
```

```python
df1.reset_index(inplace=True)
```

### Project Part 3 - Generate Map to Visualise neighbourhood

```python
# Create Map using folium 
TorontoMap = folium.Map(location=[43.65, -79.38],zoom_start = 10)
```

```python
# add markers to map
for lat, lng, label in zip(TorontoData['Latitude'], TorontoData['Longitude'], TorontoData['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(TorontoMap)  
    
TorontoMap
```
